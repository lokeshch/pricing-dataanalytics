```python
pricingReferenceSourceTableName= dbutils.widgets.get("prm_pricingReferenceSoureTableName")
pricingReferenceSinkLayerName='bronze'
pricingReferenceSinkLayerStorageAccountName='mydevstoragelake'
pricingReferenceSinkFolderName='reference-data'
```

```python
#JDBCconnectionUrl = "jdbc:sqlserver://asqludacoursesserver.database.windows.net;encrypt=true;databaseName=asqludacourses;user=sourcereader;password=DBReader@2024";
JDBCconnectionUrl=dbutils.secrets.get(scope = "azure-key-vault-scope", key = "JDBCconnectionUrlSecret")
print(JDBCconnectionUrl)
pricingReferenceSourceTableDF =(
    spark
    .read
    .format("jdbc")
    .option("url", JDBCconnectionUrl)
    .option("dbtable", pricingReferenceSourceTableName)
    .load()
)
```

```python
pricingReferenceSourceTableFolder=pricingReferenceSourceTableName.replace('.', '/')
```

```python
pricingReferenceSinkFolderPath=f"abfss://{pricingReferenceSinkLayerName}@{pricingReferenceSinkLayerStorageAccountName}.dfs.core.windows.net/{pricingReferenceSinkFolderName}/{pricingReferenceSourceTableFolder}"
(
    pricingReferenceSourceTableDF
    .write
    .mode("overwrite")
    .json(pricingReferenceSinkFolderPath)
)
```
